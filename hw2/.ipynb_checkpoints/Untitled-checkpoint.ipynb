```python
import  json, ssl, urllib.request


url = '	https://ws.yunlin.gov.tw/001/Upload/539/opendata/15369/364/0a7bfbd9-d130-4658-a27e-b85e1261ba1f.json'
context = ssl._create_unverified_context()

with urllib.request.urlopen(url, context=context) as jsondata:
    #將JSON進行UTF-8的BOM解碼，並把解碼後的資料載入JSON陣列中
     data = json.loads(jsondata.read().decode('utf-8-sig')) 
    
print (data)
```

```python
#1 台灣高中學生數折線圖

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

plt.rcParams['font.family'] = ['Microsoft JhengHei'] 

df2 = pd.DataFrame(data)
df2 = df2[(df2['項目'] == '各級學校學生數-高級中等學校') & (df2['性別'] != '...')]

#總數量
x = df2['年度']
y = df2['性別'].astype(int)
plt.plot(x, y,label='總數量',color='orange')

#男生
x = df2['年度']
y = df2['男性'].astype(int)
plt.plot(x, y,label='男生',color='blue')

#女生
x = df2['年度']
y = df2['女性'].astype(int)
plt.plot(x, y,label='女生',color='red')

plt.title('高中裸眼視力不良學生數')
plt.xlabel('年度')
plt.ylabel('數量')

plt.legend()
plt.show()
```

```python
#2 台灣高中裸眼視力不良學生數折線圖

df = pd.DataFrame(data)
df = df[(df['項目'] == '裸眼視力不良學生數-高級中等學校普通科、綜合高中') 
        & (df['性別'] != '...') & (df['年度'] != '100') & (df['年度'] != '101') & (df['年度'] != '102')]

#總數量
x = df['年度']
y = df['性別'].astype(int)
plt.plot(x, y,label='總數量',color='orange')

#男生
x = df['年度']
y = df['男性'].astype(int)
plt.plot(x, y,label='男生',color='blue')

#女生
x = df['年度']
y = df['女性'].astype(int)
plt.plot(x, y,label='女生',color='red')

plt.title('高中裸眼視力不良學生數')
plt.xlabel('年度')
plt.ylabel('數量')

plt.legend()
plt.show()
```

```python
#3 比對兩圖，繪製出高中裸眼視力不良學生佔全體學生的比例

df = pd.DataFrame(data)
df = df[(df['項目'] == '各級學校學生數-高級中等學校') & (df['性別'] != '...')]

df2 = pd.DataFrame(data)
df2 = df2[(df2['項目'] == '裸眼視力不良學生數-高級中等學校普通科、綜合高中') 
        & (df2['性別'] != '...') & (df2['年度'] != '100') & (df2['年度'] != '101') & (df2['年度'] != '102')]

merged_df = pd.merge(df, df2, on=['年度'], suffixes=('_學生總數', '_近視不良數'))
merged_df['全體近視比例'] = merged_df['性別_近視不良數'].astype(int) / merged_df['性別_學生總數'].astype(int) *100
merged_df['男生近視比例'] = merged_df['男性_近視不良數'].astype(int) / merged_df['男性_學生總數'].astype(int) *100
merged_df['女生近視比例'] = merged_df['女性_近視不良數'].astype(int) / merged_df['女性_學生總數'].astype(int) *100

#總比例
x = merged_df['年度']
y = merged_df['全體近視比例']
plt.plot(x, y,label='總數',color='orange')

#男生
x = merged_df['年度']
y = merged_df['男生近視比例']
plt.plot(x, y,label='男生',color='blue')

#女生
x = merged_df['年度']
y = merged_df['女生近視比例']
plt.plot(x, y,label='女生',color='red')

plt.title('高中裸眼視力不良學生數比例')
plt.xlabel('年度')
plt.ylabel('比例(%)')

plt.legend()
plt.show()
```
