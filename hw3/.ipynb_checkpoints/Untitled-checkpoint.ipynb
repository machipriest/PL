# 網路爬蟲(PTT電影版)

```python
#對電腦網頁版ptt提出連線請求
import urllib.request as req

url="https://www.ptt.cc/bbs/movie/index.html"

#踩雷註記：我們要裝成一個真正的使用者，所以要在標頭裡面加上一般使用者資訊
#使用者資訊：f12 > Network > Headers 拉到最下面有 user-agent

request=req.Request(url,headers={
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"
})

#開啟網頁並讀取HTML

with req.urlopen(request) as resp:
  data=resp.read().decode("utf-8")
print(data)
```

```python
#印出文章標題

from bs4 import BeautifulSoup as bs

root=bs(data,"html.parser")

#尋找特定div標籤
titles=root.find_all("div",class_="title") 

for title in titles:
    print(title.text.strip())
```

```python
#後續補充，如果不想要被刪除的標題，加上if條件檢查<a>標籤(超連結，亦即有其他子網頁)是否為空即可

titles2=root.find_all("div",class_="title") 

for title2 in titles2:  
  if title2.a!=None:
    print(title2.a.string)
    
```

```python
#一頁的資料不夠，要找"上一頁"按鈕後的網址

links = root.find_all("a", class_="btn wide")
for link in links:
    if link.text == "‹ 上頁":
        print(link.attrs["href"])
        
#對比會發現少了前面的主網址，之後進入迴圈時要加上
```

```python
#以上全部合體，產生json檔案

import json
#import urllib.request as req
import requests
from bs4 import BeautifulSoup as bs


def Main():
    main_url = "https://www.ptt.cc"
    
    #有趣的紀錄：之前我都是用%來表示，看到有個f-string可以表示，蠻有趣的
    sub_url = f"/bbs/movie/index.html"

    #準備清單
    data = list()
    
    #開始迴圈，range(100)為迴圈次數(即查100頁)
    for _ in range(100):
        
    #超級踩雷註記：由於str無法被append進list，用一開始的提取字串會不好做json，這邊改用bs4的lxml幫助我們轉換成tag
    
        #提出連線請求並讀取HTML
        all_url = f"{main_url}{sub_url}"
        #request = req.Request(all_url,headers={
            #"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"
        #})
        #with req.urlopen(request) as resp:
          #data = resp.read().decode("utf-8")


        #尋找特定div標籤
        bs_root = bs(requests.get(all_url).text, features="lxml")
        #root = bs(data,"html.parser")
        titles = bs_root.find_all("div",class_="title") 

        for title in titles:
            #踩雷註記：要先把空格字元刪掉
            title = title.text.strip()
            title = JsonTitle(title)
            if title is None:
                continue
            data.append(title)

        sub_url = NextPage(root)

    with open("Movie.json", "w", encoding="UTF-8") as file:
        json.dump(data, file, ensure_ascii=False, indent=4)



def NextPage(root):
    links = root.find_all("a", class_="btn wide")
    for link in links:
        if link.text == "‹ 上頁":
            return link.attrs["href"]
        
        
def JsonTitle(title: str):
    if "本文已被刪除" in title:
        return
    if "[" not in title:
        return
    if "]" not in title:
        return
    #因為目標是想要看到大家對電影的評價(好雷/普雷/負雷)，故增加限制
    if "雷" not in title:
        return

    left = title.index("[") + 1
    right = title.index("]")

    types = title[left:right].strip()
    title = title[right + 1 :].strip()

    return {"Type": types, "Title": title}


if __name__ == "__main__":
    Main()
```
